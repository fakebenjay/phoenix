```python
counter = 0
for index,row in df.iterrows():
    counter += 1
    
    if (counter != 5361):
        continue
        
    query = gmaps.geocode(str(row['SITE_ADDR']) + ' ' + str(row['S_CITY']) + ' AZ ' + str(row['S_ZIP']) + ' USA')
    print(str(counter) + '. ' + row['SITE_ADDR'])
    
    if 'bounds' in query[0]['geometry'].keys():
        df.at[index,'north'] = float(query[0]['geometry']['bounds']['northeast']['lat'])
        df.at[index,'south'] = float(query[0]['geometry']['bounds']['southwest']['lat'])
        df.at[index,'east'] = float(query[0]['geometry']['bounds']['northeast']['lng'])
        df.at[index,'west'] = float(query[0]['geometry']['bounds']['southwest']['lng'])
    else:
        print('NO BOUNDS')
        
    if 'location' in query[0]['geometry'].keys():
        df.at[index,'lat'] = float(query[0]['geometry']['location']['lat'])
        df.at[index,'lng'] = float(query[0]['geometry']['location']['lng'])
    else:
        print('NO LOCATION')
        
    if 'bounds' not in query[0]['geometry'].keys() and 'location' not in query[0]['geometry'].keys():
        print('NO NOTHING')
```
